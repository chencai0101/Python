### 堆叠！添加隐藏层

> 我们可以添加将某层网络的输入提供给下一层做输出，将神经网络层层堆叠起来！

```python
import numpy as np

toes = np.array([8.5, 9.5, 9.9, 9.0])           # 脚趾的数量
wlrec = np.array([0.65, 0.8, 0.8, 0.9])         # 胜负记录
fans = np.array([1.2, 1.3, 0.5, 1.0])           # 粉丝的数量

weigths_output=np.matrix([[0.1, 0.1, -0.3],      # 当前的输出层
          [0.1, 0.2, 0.0],                      
          [0.0, 1.3, 0.1]])

weigths_hid=np.matrix([[0.1, 0.2, -0.1],        # 添加一个隐藏层，代码层面就是多创建一个矩阵 
          [-0.1, 0.1, 0.9],                      
          [0.1, 0.4, 0.1]])

weigths=[weigths_hid,weigths_output]            # 将隐藏层的权重矩阵写到一起，此时应是一个三维矩阵

input = np.array([toes[0], wlrec[0], fans[0]])  # 本次被预测样本

def forward_propagation(vector, martrix):
    output = [0, 0, 0] 
    for i in range(len(vector)):
        weight = np.transpose(martrix[i,:])     # 所有行的第i个元素，即第i列元素， 转置保证满足矩阵乘法
        output[i] = np.dot(vector,weight)[0,0]  # 点乘的结果被放到了二维数组中，但是我们只需要值而不需要数组的结构。所以用[0,0] 取出    
    return output

def neural_network(input, weigths):
    hid = forward_propagation(input, weigths[0])# 计算隐藏层的输出
    pred = forward_propagation(hid, weigths[1]) # 计算输出层的输出
    return pred

pred = neural_network(input, weigths)
print(pred)
```
