### 使用多个输入与多个输出进行预测

> 这里将前两节的内容联系起来，构建全连接神经网络。下图所示网络实际上是将上一层的输入作为下一层的输出。这种行为被称为堆叠神经网络。被堆叠的部分称为隐藏层。

```python
import numpy as np

toes = np.array([8.5, 9.5, 9.9, 9.0])           # 脚趾的数量
wlrec = np.array([0.65, 0.8, 0.8, 0.9])         # 胜负记录
fans = np.array([1.2, 1.3, 0.5, 1.0])           # 粉丝的数量

weigths=np.matrix([[0.1, 0.1, -0.3],            # 以行向量存储的矩阵（编码啥样，真实矩阵就啥样）
          [0.1, 0.2, 0.0],                      # 第一行就是是否受伤节点的三个权值
          [0.0, 1.3, 0.1]])

input = np.array([toes[0], wlrec[0], fans[0]])  # 本次被预测样本

def forward_propagation(vector, martrix):
    output = [0, 0, 0] 
    for i in range(len(vector)):
        weight = np.transpose(martrix[i,:])     # 所有行的第i个元素，即第i列元素， 转置保证满足矩阵乘法
        output[i] = np.dot(vector,weight)       
    return output

def neural_network(input, weigths):
    pred = forward_propagation(input, weigths)
    return pred

pred = neural_network(input, weigths)
print(pred)
```
